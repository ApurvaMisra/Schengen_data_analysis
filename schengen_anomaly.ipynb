Importing tables from MySQL
Converting the datatypes to numeric for plotting

```python
mydb= mysql.connector.connect(host="127.0.0.1", user="root", passwd="***", database="atlasmeasure")
measurement_dict={}
measurements=["measurement_loc1","measurement_loc2","measurement_loc3","measurement_loc4","measurement_loc5","measurement_loc6","measurement_loc7","measurement_loc8","measurement_loc9","measurement_loc10","measurement_loc11","measurement_loc12","measurement_loc13","measurement_loc14"]
for measure_num in measurements:
    measurement_dict[measure_num]=pd.read_sql("SELECT * FROM "+measure_num, con=mydb)
    measurement_dict[measure_num].latitude=pd.to_numeric(measurement_dict[measure_num].latitude, downcast='float')
    measurement_dict[measure_num].longitude=pd.to_numeric(measurement_dict[measure_num].longitude, downcast='float')
    measurement_dict[measure_num]["month"] = pd.DatetimeIndex(measurement_dict[measure_num].timestamp).month
    measurement_dict[measure_num]["year"] = pd.DatetimeIndex(measurement_dict[measure_num].timestamp).year
    measurement_dict[measure_num]["day_name"] = pd.DatetimeIndex(measurement_dict[measure_num].timestamp).weekday_name
    measurement_dict[measure_num]["day"]=pd.Series()
    measurement_dict[measure_num]["hour"] = pd.DatetimeIndex(measurement_dict[measure_num].timestamp).hour
print(measurement_dict["measurement_loc1"].head(5))

'''
for index,row in measurement_loc1.iterrows():
    if (row.day_name=="Monday"):
        row.day=1
    elif (row.day_name=="Tuesday"):
        row.day=2
    elif (row.day_name=="Wednesday"):
        row.day=3
    elif (row.day_name=="Thursday"):
        row.day=4
    elif (row.day_name=="Friday"):
        row.day=5
    elif (row.day_name=="Saturday"):
        row.day=6
    elif (row.day_name=="Sunday"):
        row.day=7
'''
fil_loc1=measurement_dict["measurement_loc14"][ pd.DatetimeIndex(measurement_dict["measurement_loc14"]['timestamp']).year==2017]
print(fil_loc1.shape)
#Each unique timestamp corresponds to a particular traceroute when looking at consecutive rows
timestamps=measurement_dict["measurement_loc14"].timestamp.unique()
#print(measurement_loc1.groupby(['month']).agg(['count']))#Number of rows corresponding to each month
```

Looking for traceroutes not belonging to Schengen countries 
and also not belonging to EU countries 

```python
schengen=["AT","BE","CZ","DK","EE","FI","FR","DE","GR","HU","IS","IT","LV","LI","LT","LU","MT","NL","NO","PL","PT","SK","SI","ES","SE","CH"]
eu_countries=['AL', 'AD', 'AM', 'AT', 'BY', 'BE', 'BA', 'BG', 'CH', 'CY', 'CZ', 'DE',
        'DK', 'EE', 'ES', 'FO', 'FI', 'FR', 'GB', 'GE', 'GI', 'GR', 'HU', 'HR',
        'IE', 'IS', 'IT', 'LI', 'LT', 'LU', 'LV', 'MC', 'MK', 'MT', 'NO', 'NL', 'PL',
        'PT', 'RO', 'RU', 'SE', 'SI', 'SK', 'SM', 'TR', 'UA', 'VA']
print(fil_loc1[fil_loc1.country.isin(schengen)==False].country.unique())
other=fil_loc1[fil_loc1.country.isin(schengen)==False].country.unique()
for i in other:
    #print(i)
    print("{}:{}".format(i,len(fil_loc1[fil_loc1["country"]==i])))
```

Obtaining the percentage of traceroutes belonging to each
country every month

```python
grouped=fil_loc1.groupby(["month","country"],as_index=True).agg({"country":["count"]})
grouped.reset_index(col_level=1,inplace=True)
grouped.columns = grouped.columns.droplevel(0)
#sort grouped and plot 
grouped.sort_values(["month","count"],ascending=False,inplace=True)
grouped[grouped.month==9]
summation=grouped.groupby("month")["count"].sum()
summation
  
for i in grouped.month.unique():
    replacem=[x/summation[i] for x in grouped[grouped.month==i]["count"]]
    mask=grouped.month==i
    grouped.loc[mask,"count"]=replacem
    
grouped=grouped[grouped.country.isin(schengen)==False]
grouped_count=grouped.pivot(index="month", columns="country", values="count")
other_countries=grouped_count.columns
eu_selection=list(map(lambda x: x in eu_countries, grouped.country))
#print(eu_selection)
grouped["eu"]=eu_selection
```

Plotting percentage traceroutes for each country for each month in 
the year 2017 and also differentiating between EU and non-EU countries. It shows GB to be the highest for this particular target

```python
ax=sns.lineplot(x="month",y="count",hue="country",style="eu",data=grouped)
ax.set(xlabel="months(2017)", ylabel="percentage ratio country visited per month")
```

Obtaining the percentage of traceroutes belonging to each
country for each day type

```python
grouped_day=fil_loc1.groupby(["day_name","country"],as_index=True).agg({"country":["count"]})
grouped_day.reset_index(col_level=1,inplace=True)
grouped_day.columns = grouped_day.columns.droplevel(0)
grouped_day.sort_values(["day_name","count"],ascending=False,inplace=True)
summation_day=grouped_day.groupby("day_name")["count"].sum()
summation_day

for k in grouped_day.day_name.unique():
    replace=[x/summation_day[k] for x in grouped_day[grouped_day.day_name==k]["count"]]
    mask=grouped_day.day_name==k
    grouped_day.loc[mask,"count"]=replace
    
grouped_day=grouped_day[grouped_day.country.isin(schengen)==False]
eu_selection=list(map(lambda x: x in eu_countries, grouped_day.country))
grouped_day["eu"]=eu_selection
```

Percentage traceroutes for each day for a particular country shows

```python
ax=sns.lineplot(x="day_name",y="count",hue="country",style="eu",data=grouped_day)
ax.set(xlabel="Days(2017)", ylabel="percentage ratio country visited each day type")
```

Obtaining the percentage of traceroutes belonging to each
country for each of the 24 hours

```python
grouped_hour=fil_loc1.groupby(["hour","country"],as_index=True).agg({"country":["count"]})
grouped_hour.reset_index(col_level=1,inplace=True)
grouped_hour.columns = grouped_hour.columns.droplevel(0)
grouped_hour.sort_values(["hour","count"],ascending=False,inplace=True)
summation_hour=grouped_hour.groupby("hour")["count"].sum()
summation_hour


for j in grouped_hour.hour.unique():
    replace=[x/summation_hour[j] for x in grouped_hour[grouped_hour.hour==j]["count"]]
    mask=grouped_hour.hour==j
    grouped_hour.loc[mask,"count"]=replace
grouped_hour=grouped_hour[grouped_hour.country.isin(schengen)==False]
eu_selection=list(map(lambda x: x in eu_countries, grouped_hour.country))
grouped_hour["eu"]=eu_selection
```

Percentage traceroutes for each hour for a particular country shows
that a larger number of traceroutes were conducted in the early hours
of the day which might also depend on when Atlas decides to conduct 
traceroutes for its users 

```python
ax=sns.lineplot(x="hour",y="count",hue="country",style="eu",data=grouped_hour)
ax.set(xlabel="Hours of day(2017)", ylabel="percentage ratio country visited each hour type")
```

Correlation matrix for countries based on numbers from each day

```python
corr_days=grouped_day_count.corr(method='pearson')
mask=np.zeros_like(corr_days)
mask[np.triu_indices_from(mask)]=True
corr_days
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr_days,cmap=cmap,mask=mask, xticklabels=True,yticklabels=True, linewidth=0.01,square=True,ax=ax)
plt.yticks(rotation = 45)
plt.xticks(rotation = 45)
plt.show()
```
